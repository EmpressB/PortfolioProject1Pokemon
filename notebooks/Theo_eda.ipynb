```python
import pandas as pd 
import matplotlib.pyplot as plt 
import statistics as stat 
```

```python
df = pd.read_csv('../data/raw/pokemon_data.csv')
```

```python
#Read the Pokemon CSV File
with open('../data/raw/pokemon_data.csv', 'r') as file:
    lines = file.readlines()

for line in lines:
    print(line.strip())
```

```python
#Getting the highest attack power Pokemon
highest_attack = 0
strongest_pokemon = ""

for line in lines[1:]:
    data = line.strip().split(',')

    name = data[1]
    attack = int(data[6])

    if attack > highest_attack:
        highest_attack = attack
        strongest_pokemon = name
```

```python
#Print Result
print("Pokemon with highest attack:", strongest_pokemon)
print("Attack Power:", highest_attack)
```

```python
#Getting the lowest attack power Pokemon
lowest_attack = float('inf')
weakest_pokemon = ""

for line in lines[1:]:
    data = line.strip().split(',')

    name = data[1]
    attack = int(data[6])

    if attack < lowest_attack:
        lowest_attack = attack
        weakest_pokemon = name
```

```python
#Print Result 
print("Pokemon with the lowest attack:", weakest_pokemon)
print("Attack Power:", lowest_attack)
```

```python
# Create bar graph
x = range(len(pokemon_names))
width = 0.4

plt.figure(figsize=(8,6))

plt.bar(x, attack_stats, width=width, label='Attack')
plt.bar([i + width for i in x], defense_stats, width=width, label='Defense')

plt.title("Strongest vs Weakest Pokemon Stats")
plt.xlabel("Pokemon")
plt.ylabel("Stat Value")
plt.xticks([i + width/2 for i in x], pokemon_names)
plt.legend()

plt.show()
```
