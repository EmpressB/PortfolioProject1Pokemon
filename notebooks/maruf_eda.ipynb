## EDA Question: Which Pokemon type has the highest average base stat total?

My hypothesis is that Dragon type Pokemon have the highest average base stat total because they are typically rare and legendary-level Pokemon with extremely high stats compared to common types like Bug or Normal.

## Imports
Loading the libraries needed for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data
Reading the Pokemon CSV file into a pandas DataFrame.

```python
df = pd.read_csv('../data/raw/pokemon_data.csv')
df.head()
```

## Analysis
Grouping Pokemon by their primary type and calculating the average base stat total for each type.

```python
avg_by_type = df.groupby('type_1')['base_stat_total'].mean().sort_values(ascending=False)
print(avg_by_type)
```

```python
avg_by_type.plot(kind='bar', figsize=(12,6), color='skyblue')
plt.title('Average Base Stat Total by Type')
plt.xlabel('Type')
plt.ylabel('Average Base Stat Total')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Visualization
Bar chart showing the average base stat total for each Pokemon type, sorted from highest to lowest.

## Conclusion

My hypothesis was that Dragon type Pokemon would have the highest average base stat total. The data confirms this. Dragon types average 490 base stat total, significantly higher than any other type. Bug type came in last at 375.

This makes sense because Dragon type Pokemon are typically rare, hard to catch, and often legendary. Games are designed to make them feel powerful, and the stats reflect that.
