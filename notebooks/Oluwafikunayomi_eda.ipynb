```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stat
```

```python
df = pd.read_csv("../data/raw/pokemon_data.csv")
```

# Which Pokémon type appears most frequently?

```python
df["type_1"].head()
```

```python
type_counts = df["type_1"].value_counts()

print(type_counts)
```

```python
most_common_type = df["type_1"].value_counts().idxmax()

print("Most common Pokémon type:", most_common_type)
```

```python
most_common_count = df["type_1"].value_counts().max()

print("Number of Pokémon:", most_common_count)
```

```python
type_counts.plot(kind="bar")

plt.title("Frequency of Pokémon Types")
plt.xlabel("Pokémon Type")
plt.ylabel("Count")

plt.show()
```
